### Hypothesis testing

Determine whether ice cream reviews significantly change the proportion positive reviews. We will use two proportion z test to determine the significance 

$H_{0}$: *$P_{1}$* = *$P_{2}$*             $H_{1}$: *$P_{1}$* $\ne$ *$P_{2}$* 

Where alpha is $\alpha$ = 0.05

1\. Retrieve data

```python
import pandas as pd

data = pd.read_csv('./kinton_reviews.csv')

ice_cream = data[284:]

# keep only data from after the ice cream promotion
no_ice = data[0:283]
```

2\. Determine each proportion

```python
ice_cream_good = len(ice_cream[ice_cream['review_rating'] >= 4])

no_ice_good = len(no_ice[no_ice['review_rating'] >= 4])


ice_cream_prop = ice_cream_good / len(ice_cream)

no_ice_prop = no_ice_good / len(no_ice)

total_prop = (ice_cream_good + no_ice_good) / (len(ice_cream) + len(no_ice))

print("ice cream reviews: ", ice_cream_prop,"\nno ice cream reviews: ", no_ice_prop, "\ntotal proportion: ", total_prop)
print(len(ice_cream))
print(len(no_ice))
```

3\. Calculate Z 

```python
import math

num = ((ice_cream_prop - no_ice_prop) - 0)

den = math.sqrt((total_prop*(1-total_prop)) * ((1/len(ice_cream)) + (1/len(no_ice))))

z = num / den

z
```

4\. Compare with $\alpha$/2

Since we are doing a two-tailed test, we must double the p-value to 0.1.

```python
confidence = .99992

p_val = (1 - .99992) * 2

print(p_val)
```
